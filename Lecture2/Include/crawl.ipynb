```python
def scroll():
    #페이지의 맨 끝에 다다랐는지 확인하고 그렇지 않다면 계속해서 end 키를 누르는 코드.
    elem = driver.find_element(By.TAG_NAME, "body")

    #스크롤길이정보 획득
    last_height = driver.execute_script("return document.body.scrollHeight")

    #계속해서반복
    while True:
            #end키 3번 누르기
        for i in range(3):
            elem.send_keys(Keys.END)
            time.sleep(0.2)
        
            #이번 스크롤 길이정보 획득
        new_height = driver.execute_script("return document.body.scrollHeight")

            #만약 새 스크롤 길이정보가 이전과 달라지지 않았다면
        if new_height == last_height:
            print("********" + "break********")
                    #루프 종료
            return
            #직전 스크롤 길이정보에 이번 스크롤 길이정보값을 할당
        last_height = new_height
```

```python
def cal_satis(li):
    for i, elem in enumerate(li):
        if (elem.get_attribute('src').split('/')[-1] != 'orange-pow.svg'):
            return i
    return 5
```

```python
driver.get('https://klue.kr/search?query=ACADEMIC%20ENGLISH%E2%85%A1&sort=year_term')
scroll()
```

```python
#접근할 각 강의평가 목록을 가져옵니다
elements = driver.find_elements(By.XPATH, '//ul/div/div/li')
#각 정보를 담을 리스트를 생성합니다
links, 학기, 학수번호, 교수명, 강의시간 = [], [],[],[],[]
학습량, 난이도, 강의력, 성취감, 강의만족도 = [],[],[],[],[]

#for문을 사용하여 elements의 각 element를 탐색합니다.
for i, element in enumerate(elements):
    #평가정보 div의 윗단 div를 top으로 선언
    links.append(element.find_element(By.XPATH, 'a').get_attribute('href'))
    top =  element.find_element(By.XPATH, 'a/div[1]')
    #top에서 학기와 학수번호 정보를 가져옵니다.(2019년 2학기 - KECE208(03))
    li = top.find_element(By.XPATH, 'p[1]').text.split(' - ') #정보가 ' - '로 나뉘어있으므로, 해당 값을 기준으로 정보를 분할합니다.

    #각 정보를 필요한 리스트에 담습니다.
    학기.append(li[0])
    학수번호.append(li[1])
    
    #평가정보 div의 아랫단 div를 bottom으로 선언
    bottom = element.find_element(By.XPATH, 'a/div[2]')

    #bottom에서 교수명을 가져와 리스트에 담습니다.
    교수명.append(bottom.find_element(By.XPATH, 'div[1]/p[2]/span').text)
    강의시간.append(bottom.find_element(By.XPATH, 'div[1]/p[3]').text.split('#')[1])

    #오래된 강의의 강의평에 대해서는 학습량, 난이도, 강의력, 성취감 정보가 존재하지 않는 경우가 있습니다.
    #존재하지 않는 영역의 값을 가져올땐 오류가 발생하므로, try-except문을 사용하여 오류처리를 해줍니다.
    try:
        학습량.append(bottom.find_element(By.XPATH, 'div[2]/div/div[1]/span[2]').text)
        난이도.append(bottom.find_element(By.XPATH, 'div[2]/div/div[2]/span[2]').text)
        강의력.append(bottom.find_element(By.XPATH, 'div[2]/div/div[3]/span[2]').text)
        성취감.append(bottom.find_element(By.XPATH, 'div[2]/div/div[4]/span[2]').text)
    except:
        #강의평가 요약을 지원하지 않는 강의일 경우
        print("강의평가 요약을 지원하지 않는 강의입니다")
        학습량.append('-')
        난이도.append('-')
        강의력.append('-')
        성취감.append('-')

    #bottom에서 강의만족도를 가져와 리스트에 담습니다.
    강의만족도.append(bottom.find_element(By.XPATH, 'div[3]/div/span').text)
    
    #가져온 값들을 출력합니다.
    print(f"학기: {학기[i]} 학수번호: {교수명[i]} 교수명: {교수명[i]} 강의시간: {강의시간[i]}")
    print(f"학습량: {학습량[i]}, 난이도: {난이도[i]}, 강의력: {강의력[i]}, 성취감: {성취감[i]} 강의만족도: {강의만족도[i]}")
    print(links[i])
```

```python
import pandas as pd
info = pd.DataFrame((zip(학기, 학수번호, 교수명, 강의시간, 학습량, 난이도, 강의력, 성취감, 강의만족도)), columns = ["학기", "학수번호", "교수명", "강의시간", "학습량", "난이도", "강의력", "성취감", "강의만족도"])
info
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

#각 정보를 담을 리스트를 생성합니다
만족도, 학습량, 난이도, 강의력, 성취감 = [],[],[],[],[]
user_id, 강의평, lecture_id = [],[],[]
#각 강의평가 페이지를 순회합니다.
for i, link in enumerate(links):
    driver.get(link)
    try:
        # 페이지가 로드될 때까지 대기합니다 (최대 10초)
        WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, '//section[3]/div/div/div[2]/div/div')))
    except TimeoutException:
        print(f"Element not found in {link}")
        continue
    #접근할 각 강의평가 목록을 가져옵니다
    elements = driver.find_elements(By.XPATH, '//section[3]/div/div/div[2]/div/div')
    print(f"{i}, {link}")
    #for문을 사용하여 elements의 각 element를 탐색합니다.
    for j, element in enumerate(elements):
        top = element.find_element(By.XPATH, 'div/div/div[1]')
        #강의평 데이터를 수집합니다.
        만족도.append(cal_satis(top.find_elements(By.XPATH, 'div/div[1]/div/img')))
        학습량.append(int(top.find_element(By.XPATH, 'div/div[2]/div/div[1]/span').text))
        난이도.append(int(top.find_element(By.XPATH, 'div/div[2]/div/div[2]/span').text))
        강의력.append(int(top.find_element(By.XPATH, 'div/div[2]/div/div[3]/span').text))
        성취감.append(int(top.find_element(By.XPATH, 'div/div[2]/div/div[4]/span').text))
        user_id.append(top.find_element(By.XPATH, 'div[2]/div/span').text)
        강의평.append(element.find_element(By.XPATH, 'div/div/div[2]/div').text)
        #강의 id를 추가합니다.
        lecture_id.append(i)
        print(j)
        print(f"만족도 {만족도[-1]}, 유저 {user_id[-1]}, 강의평 {강의평[-1]}, lecture_id {lecture_id[-1]}")
```

```python
def codify(data):
    year = data.split(" ")[0][2:4]
    if data.split(" ")[1] == "1학기":
        semaster = "1"
    elif data.split(" ")[1] == "여름학기":
        semaster = "2"
    elif data.split(" ")[1] == "2학기":
        semaster = "3"
    elif data.split(" ")[1] == "겨울학기":
        semaster = "4"
    return int(year+semaster)
```

```python
학기 = list(map(codify, 학기))
학기
```

```python
info["학기"] = 학기
info
```

```python
def sliceing(data):
    return data.split("(")[1].replace(")","")
```

```python
학수번호 = list(info["학수번호"])
학수번호 = list(map(sliceing, 학수번호))
info["학수번호"] = 학수번호
info
```

```python
def timing(data):
    data = data[:-1]
    li = data.split(" ")
    day = ""
    for i in li:
        day = f"{day}{i.split('(')[0]}, "
    time = int(li[0].split("(")[1][0])
    return day[:-2], time
강의시간 = list(info["강의시간"])
강의요일, 강의시간 = zip(*map(timing, 강의시간))
강의시간
info["강의요일"] = 강의요일
info["강의시간"] = 강의시간
info
```
